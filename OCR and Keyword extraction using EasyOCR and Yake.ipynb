### Importing libraries

```python
import easyocr
import yake 
from wordcloud import WordCloud 
import matplotlib.pyplot as plt
import cv2
```

### Case I

#### Plotting the image

```python
img = cv2.imread(r"C:\Users\imane\Downloads\news_1.png")
plt.imshow(img)
plt.show()
```

#### Extracting the text

```python
reader = easyocr.Reader(['en'])
results = reader.readtext(r"C:\Users\imane\Downloads\news_1.png", detail=0)
results
```

In the above I plotted the image from which I want to extract text using cv2, and then extracted the text contained within it using EasyOCR which is pretty straightforward to implement.

```python
text = ' '.join(l for l in results)
text
```

I then converted the extracted text to a string using the join method.

#### Extracting and plotting keywords

```python
yake = yake.KeywordExtractor() 
keywords = yake.extract_keywords(text)
keywords = [kw for kw, _ in keywords]
print(keywords[:5])
```

```python
wordcloud = WordCloud().generate(' '.join(keywords)) 
  
# Display the WordCloud 
plt.figure(figsize=(10,10)) 
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis('off') 
plt.show()
```

And finally I extracted the keywords and used WordCloud to plot the most important keywords i.e the ones with the highest probabilities.
